```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
import mysql.connector

# Function to create a MySQL database if it doesn't exist
def create_database_if_not_exists(user, password, host, database_name):
    # Connect to MySQL server (without specifying a database)
    connection = mysql.connector.connect(user=user, password=password, host=host)
    cursor = connection.cursor()
    
    # Create the database if it doesn't exist
    cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database_name}")
    connection.commit()
    
    # Close the connection
    cursor.close()
    connection.close()
    print(f"Database '{database_name}' checked/created.")

# Function to load a dataframe into a MySQL table
def load_dataframe(df, table_name, engine):
    df.to_sql(table_name, con=engine, if_exists='replace', index=False)
    print(f"Data loaded into table '{table_name}'.")

# Step 1: Define MySQL credentials and database name
user = 'root'
password = 'S@urabh13'
host = 'localhost'
database_name = 'yelp'


# Step 2: Create the database if it doesn't exist
create_database_if_not_exists(user, password, host, database_name)

# Step 3: Create an SQLAlchemy engine to connect to the MySQL database
# Replace user, password, and host with your MySQL credentials
engine = create_engine(f'mysql+pymysql://root:S%40urabh13@localhost/yelp')


# Step 5: Load each dataframe into its own MySQL table
load_dataframe(business_df, 'business', engine)
load_dataframe(review_df, 'review', engine)
#load_dataframe(user_df, 'user', engine)
#load_dataframe(tip_df, 'tip', engine)
#load_dataframe(checkin_df, 'checkin', engine)
```

```python
from sqlalchemy import text

# Ensure the column is TEXT
alter_query = text("ALTER TABLE checkin MODIFY COLUMN date TEXT;")

# Execute the alter query
with engine.connect() as connection:
    connection.execute(alter_query)


# Preprocess to handle long date strings
def preprocess_checkin_df(checkin_df):
    new_rows = []
    
    for _, row in checkin_df.iterrows():
        # Split the concatenated dates into individual dates
        dates = row['date'].split(', ')
        
        # Create a new row for each date
        for date in dates:
            new_rows.append({'business_id': row['business_id'], 'date': date})
    
    # Convert the new rows into a DataFrame
    return pd.DataFrame(new_rows)

# Preprocess the checkin DataFrame
checkin_df_processed = preprocess_checkin_df(checkin_df)

# Load the processed DataFrame into the checkin table
load_dataframe(checkin_df_processed, 'checkin', engine)
```
